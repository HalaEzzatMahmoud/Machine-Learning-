```python
from pandas import read_csv 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
path = r"D:/GP/Cousres/ML/lecture4/task/HW/Uni-ML00-Lec4-Practice/Uni-ML00-Lec4-Practice/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width']
dataset = read_csv(path, names = names )
```

```python
x = dataset.iloc[:,[2,3]].values
#print(x)
```

```python
K_Model = KMeans(n_clusters = 3,random_state = 1)
Y_KMeans = K_Model.fit_predict(x)
print(Y_KMeans)
print("******Task1 DONE********")
```

```python
path2 = r"D:/GP/Cousres/ML/lecture4/task/HW/Uni-ML00-Lec4-Practice/Uni-ML00-Lec4-Practice/sport.csv"
names2 = ['height', 'weight', 'age']
dataset2 = read_csv(path2, names = names2)
```

```python
x2 = dataset2.iloc[:,[0,1]].values
#print(x)
```

```python
K2_Model = KMeans(n_clusters = 3,random_state = 1)
Y2_KMeans = K2_Model.fit_predict(x2)

print(Y2_KMeans)
```

```python
plt.scatter(x2[Y2_KMeans == 0,0],x2[Y2_KMeans == 0,1],c = 'red',label ='Cluster 1')
plt.scatter(x2[Y2_KMeans == 1,0],x2[Y2_KMeans == 1,1],c = 'green',label ='Cluster 2')
plt.scatter(x2[Y2_KMeans == 2,0],x2[Y2_KMeans == 2,1],c = 'yellow',label ='Cluster 3')
plt.title('Grouping Sports')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
arr1 = np.column_stack((x2, Y2_KMeans))
np.savetxt('sports2.csv',arr1,delimiter= ',',fmt =' %d')
print("File Saved")
```
