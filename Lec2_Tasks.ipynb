```python
from pandas import read_csv 
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
path = r"D:/GP/Cousres/ML/lecture2/task/UNI_ML00_Lec2_Practice/weather.csv"
dataset = read_csv(path)
```

```python
humidity9am = dataset.values[:,3]
humidity3pm = dataset.values[:,4]
pressure9am = dataset.values[:,5]
pressure3pm = dataset.values[:,6]
temp9am = dataset.values[:,7]
temp3pm = dataset.values[:,8]
corr1 = np.corrcoef(humidity9am,humidity3pm)
corr2 = np.corrcoef(pressure9am,pressure3pm)
corr3 = np.corrcoef(temp9am,temp3pm)
```

```python
print("*****corr1***","\n",corr1)
print("*****corr2***","\n",corr2)
print("*****corr3***","\n",corr3)
```

```python
print("The correlation between pressure9am & pressure3pm is the highest","\n",corr2)
```

```python
x_trainPre9am,x_testPre9am,y_trainPre3pm,y_testPre3pm = train_test_split(pressure9am,pressure3pm,test_size = 0.2,random_state = 1)
print("Xtarin \n The Xtrain size:",x_trainPre9am.size,"\n",x_trainPre9am)
print("xtest \n The Xtest size: ",x_testPre9am.size,"\n",x_testPre9am)
print("ytarin \n The Ytrain size: ",y_trainPre3pm.size,"\n",y_trainPre3pm)
print("ytest \n The Ytest size:",y_testPre3pm.size,"\n",y_testPre3pm)
```

```python
lin_gre = LinearRegression()
lin_gre.fit(x_trainPre9am.reshape(-1,1),y_trainPre3pm.reshape(-1,1))
y_prepredicted3pm = lin_gre.predict(x_testPre9am.reshape(-1,1))
#print(y_predicted3pm,y_testPre3pm)
```

```python
pyplot.scatter(pressure9am,pressure3pm)
pyplot.plot(x_testPre9am,y_prepredicted3pm,color = 'r')
pyplot.show()
```

```python
meansquresrrorPre = mean_squared_error(y_test3pm,y_predicted3pm)
r2_Pre = r2_score(y_test3pm,y_predicted3pm)
print("Mean Squred Error for pressure: ",meansquresrrorPre,"\n R2 Score for pressure: ",r2_Pre)
```

```python
slopePre = lin_gre.coef_
point_yPre = lin_gre.intercept_
print("Slope for pressure: ",slopePre,"\n Point With Y for pressure: ",point_yPre)
```
