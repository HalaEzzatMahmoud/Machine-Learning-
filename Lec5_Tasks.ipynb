```python
from pandas import read_csv 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LinearRegression
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.feature_selection import SelectKBest,chi2
```

```python
path = r"D:/GP/Cousres/ML/lecture5/task/Uni-ML00-Lec5-Practice/Uni-ML00-Lec5-Practice/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = read_csv(path, names=names )
```

```python
x = dataset.iloc[:,0:2].values
x1 = dataset.iloc[:,2:4].values
y = dataset.iloc[:,4].values
```

```python
pca= PCA(n_components=1)
f= pca.fit(x)
print(f.components_)
x2 = pca.fit_transform(x)
print(x2)
```

```python
path = r"D:/GP/Cousres/ML/lecture5/task/HW/MySolution/iris4.csv"
names = ['sepal', 'petal', 'class']
dataset = read_csv(path, names=names )
```

```python
x = dataset.iloc[:,[0,1]].values
y = dataset.iloc[:,2].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
cls = SVC()
cls.fit(x_train,y_train)
y_pred = cls.predict(x_test)
print("The confusion matrix \n " ,confusion_matrix(y_test,y_pred))
print("The classification report \n ",classification_report(y_test,y_pred))
```

```python
sepal = input("Enter sepal  : ")
petal = input("Enter petal  : ")
test = cls.predict([[sepal,petal]])
print("The predicted class is :", test[0])
```

```python
path = r"D:/GP/Cousres/ML/lecture5/task/Uni-ML00-Lec5-Practice/Uni-ML00-Lec5-Practice/Sport4.csv"
names = ['height', 'weight', 'age','class']
dataset = read_csv(path, names = names)
```
